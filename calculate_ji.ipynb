```python
import pandas as pd
from textdistance import jaccard
```

```python
data = pd.read_table("data\\imdb-title-cast-rating-desc.tsv",usecols=['title','description','genre','cast'])
data = data.dropna()
genres = data.set_index('title').to_dict()['genre']
cast = data.set_index('title').to_dict()['cast']
data = data.set_index('title').to_dict()['description']
```

```python
def jdo(row):
    try:
        return jaccard(data[row['t1']],data[row['t2']])
    except:
        return 0
```

```python
def common_genres(row):
    try:
        g1 = set(genres[row["t1"]].lower().split(","))
        g2 = set(genres[row["t2"]].lower().split(","))
        return len(g1.intersection(g2))
    except:
        return 0
```

```python
def common_cast(row):
    try:
        c1 = set(cast[row["t1"]].lower().split(","))
        c2 = set(cast[row["t2"]].lower().split(","))
        return len(c1.intersection(c2))
    except:
        return 0
```

```python
all_items = []
for chunk in pd.read_csv("pairs.csv",delimiter=" , ",names=['t1','t2'],engine="python",chunksize=100000):
    print(chunk.shape)
    chunk["ji"] = chunk.apply(lambda row: jdo(row),axis=1)
    chunk["cg"] = chunk.apply(lambda row: common_genres(row),axis=1)
    chunk["cc"] = chunk.apply(lambda row: common_cast(row),axis=1)
    chunk.to_csv('compiled.csv', mode='a', header=False)
    # print(chunk.head())
```

```python
titles = []

for chunk in pd.read_csv("pairs.csv",delimiter=" , ",names=['t1','t2'],engine="python",chunksize=100000):
    titles.extend(list(chunk["t1"]))
    titles.extend(list(chunk["t2"]))
    
```
