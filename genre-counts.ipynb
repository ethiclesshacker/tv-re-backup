```python
import pandas as pd
```

```python
data = pd.read_table("data\\imdb-title-cast-rating-desc.tsv")
data = data.dropna()
data = data[data["genre"]!="\\N"]
```

```python
data
```

```python
data = data.sample(n=5000)
```

```python
show_genres = list(data["genre"])

genres = ",".join(show_genres)
genres = genres.split(",")
genres = set(genres)
```

```python
print(genres)
print(len(genres))
```

```python
mega_set = set()
for genre in genres:
    print(genre)
    df = data[data["genre"].str.contains(genre)]
    titles = df["title"]

    for t1 in titles:
        for t2 in titles:
            if t1 != t2:
                mega_set.add(" , ".join(sorted([t1,t2])))
```

```python
print(len(mega_set))
```

```python
lil = []
while True:
    try:
        lil.append(mega_set.pop()+"\n")
        l = len(lil)
    except:
        print("Set empty finally!!")
        print(len(lil))
        break
```
